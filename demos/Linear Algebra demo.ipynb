```python
import numpy as np
```

```python
v1 = np.array([1,2,3])
v2 = np.array([4,5,6])
```

```python
print (v1)
print (v2)
```

```python
v_add = v1+v2
v_sub = v1-v2
print(v_add)
print(v_sub)
```

```python
scalar = 2
v_scaled = scalar*v1
print(v_scaled)
```

```python
dot_product = np.dot(v1,v2)
print(dot_product)
```

```python
magnitude_v1 = np.linalg.norm(v1)
print("Magnitude of v1: ",magnitude_v1)
```

```python
A = np.array([[1,2,3],[4,5,6]])
print(A)
B = np.array([[7,8,9],[10,11,12],[13,14,15]])
print(B)
```

```python
C = np.array([[1,3,4],[4,5,6]])
matrix_add = A + C
matrix_sub = A - C
print("A+C = \n",matrix_add)
print("A-C = \n",matrix_sub)
```

```python
matrix_mul = np.dot(A, B)
print(matrix_mul)
```

```python
transpose_A = A.T
print(transpose_A)
```

```python
A = np.array([[2,3],[4,-1]])
b = np.array([5,3])
x = np.linalg.solve(A,b)
print("Solution (x,y): ",x)
```

```python
A = np.array([[2,1],[1,2]])
eigenvalues, eigenvectors = np.linalg.eig(A)
print("Eigenvalues:", eigenvalues)
print("Eigenvectors:\n", eigenvectors)
```

```python
A = np.array([[4,7],[2,6]])
A_inv = np.linalg.inv(A)
print("Inverse of A:\n",A_inv)
identity = np.dot(A, A_inv)
print("A * A_inv:\n", identity)
```
