```python
def newton_raphson(g, dg, x0, tol=1e-6, max_iter=50):
    x = x0


    for i in range(max_iter):
        x_new = x - g(x) / dg(x)

        if abs(x_new - x) < tol:
            return x_new, i + 1
        x = x_new
        print(x)
    return None, max_iter

# Example: Solving x^2 - 2 = 0
g = lambda x: x**2 - 2
dg = lambda x: 2*x
root, iterations = newton_raphson(g, dg, x0=1)
print(f"Root: {root}, Iterations: {iterations}")
```
