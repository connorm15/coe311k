```python
import numpy as np

#Data points
X = np.array([1, 2, 3, 4, 5])
Y = np.array([1.2, 2.8, 3.6, 4.5, 5.1])
#stack the design matrix with a column of ones (for the intercept)
A = np.vstack([X, np.ones(len(X))]).T
#solce using least sqares
m, c = np.linalg.lstsq(A, Y, rcond=None) [0]
print(f"Slope: {m}, Intercept: {c}")
```

```python
import matplotlib.pyplot as plt
plt.scatter(X, Y, label='Data points')
plt.plot(X, m*X + c, '-g', label='Best fit line: y=mx+c')
plt.title('Least Squares Regression Fit')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```
