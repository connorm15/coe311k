```python
import random
#function to approximate Pi using the Monte Carlo Method
def approxPi(numSamples):
    counter = 0
    for s in range(0, numSamples):
        x = random.uniform(-1,1)
        y = random.uniform(-1,1)
        if(((x*x) + (y*y)) < 1):
            counter += 1
    return 4 * counter / numSamples
```

```python
import numpy as np
#creates vector of sample sizes with step size and empty vector for error results
sample_sizes = np.arange(10, 10001, 20)
errorVals = []
#calculates error for each sample size and adds result to vector
for n in sample_sizes:
    approx = approxPi(n)
    error = abs(approx - np.pi)
    errorVals.append(error)
```

```python
#plots the results of the calculated absolute error vs sample size
import matplotlib.pyplot as plt
plt.plot(sample_sizes, errorVals, 'b-')
plt.title('Error vs Sample Size')
plt.xlabel('Sample Size')
plt.ylabel('Absolute Error')
plt.show()
```
