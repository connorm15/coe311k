```python
import numpy as np
import math
#function to evaluate a value for n number of terms of the taylor series of sin(x)
def taylor(n, x):
    approximation = 0
    for i in range(n):
        term = ((-1)**i) * (x**(2*i + 1)) / math.factorial(2*i  + 1)
        approximation += term
    return approximation
```

```python
#creates array of x values, evaluations of sin for them, and empty array to store error values
x_values = np.arange(0, 10.1, 0.1)
exact_values = np.sin(x_values)
errors = {}
#approximates sin values for x using increasing number of terms, stores error to array for each value
for n in [1, 2, 3, 4]:
    approx = [taylor(n, val) for val in x_values]
    errors[n] = np.abs(np.array(approx) - exact_values)
```

```python
#plots the absolute error found above
import matplotlib.pyplot as plt
plt.figure()
for n, error_vals in errors.items():
    plt.plot(x_values, error_vals, label=f'{n} Terms')
plt.yscale('log')
plt.xlabel('x values')
plt.ylabel('Absolute Error')
plt.title('Truncation error of taylor series for different number of terms')
plt.legend()
plt.show()
```

```python
#The error increases at an exponential rate as the values of x increase
#The error behaves diferently for smaller x values, at first increasing, then decreasing, and finally increasing exponentially fast at large values
#As x-values continue to increase the reate at which the error increases decreases
```
