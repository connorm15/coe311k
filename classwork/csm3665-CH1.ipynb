```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.subplot(2, 2, 1)
x = np.linspace(0, 2*np.pi, 50)
plt.plot(x, np.sin(x), 'r')
plt.subplot(2, 2, 2)
y = np.sin(x)
plt.scatter(x,y)
plt.subplot(2, 2, 3)
x = np.random.rand(1000)
y = np.random.rand(1000)
size = np.random.rand(1000) * 50
color = np.random.rand(1000)
plt.scatter(x, y, size, color)
plt.colorbar()
plt.subplot(2, 2, 4)
plt.hist(x, 50)
plt.show()
```
