```python
import numpy as np
import random
def approxPi(numSamples):
    counter = 0
    for s in range(0, numSamples):
        x = random.uniform(-1,1)
        y = random.uniform(-1,1)
        if(((x*x) + (y*y)) < 1):
            counter += 1
    return 4 * counter / numSamples
```

```python
xVals = np.arange(0,100)
absErrorVals = []
relErrorVals = []
for x in range(len(xVals)):
    temp = approxPi(1000)
    absError = abs(np.pi - temp)
    relError = abs((np.pi - temp)/np.pi)
    absErrorVals.append(absError)
    relErrorVals.append(relError)
    
```

```python
import matplotlib.pyplot as plt
plt.plot(xVals, absErrorVals, 'r-', label='Absolute Error')
plt.plot(xVals, relErrorVals, 'b-', label='Relative Error')
plt.legend()
plt.show()
```
