```python
#Part 1
import numpy as np
# Data points 
X = np.array([1, 2, 3, 4, 5])
Y = np.array([1.2, 2.8, 3.6, 4.5, 5.1])
# Stack the design matrix with a column of ones (for the intercept)
A = np.vstack([X, np.ones(len(X))]).T

# Solve using least squares
m, c = np.linalg.lstsq(A, Y, rcond=None)[0]
print(f"Slope: {m}, Intercept: {c}")
```

```python
import matplotlib.pyplot as plt
#graph data points and least squares regression fit for them
plt.scatter(X, Y, label='Data points')
plt.plot(X, m*X + c, '-g', label='Best fit line: y=mx+c')
plt.title('Least Squares Regression Fit')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

```python
#Part 2
#Create random vectors for X and Y and run least squares regression using them
X = np.sort(np.random.uniform(1, 10, 5))
Y = np.random.uniform(1, 10, 5)
A = np.vstack([X, np.ones(len(X))]).T
m, c = np.linalg.lstsq(A, Y, rcond=None)[0]
print(f"Slope: {m}, Intercept: {c}")

#graph randomized points and least squares regression fit for them
plt.scatter(X, Y, label='Data points')
plt.plot(X, m*X + c, '-g', label='Best fit line: y=mx+c')
plt.title('Least Squares Regression Fit')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```
