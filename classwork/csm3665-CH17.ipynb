```python
import numpy as np
import matplotlib.pyplot as plt

# Base parameters
x_start = 0
x_end = 10
y_0 = 1.0
v_0 = 0.0

# h values from the challenge plus the k=10 case
configs = [(.01, 1.0), (.02, 1.0), (.05, 1.0), (.1, 1.0), (.2, 1.0), (.3, 1.0), (.5, 1.0), (.01, 10.0)]

plt.figure(figsize=(14, 7))

for h, k in configs:
    x_values = np.arange(x_start, x_end + h, h)
    y_values = np.zeros(len(x_values))
    v_values = np.zeros(len(x_values))
    
    y_values[0] = y_0
    v_values[0] = v_0

    for i in range(1, len(x_values)):
        # Euler's Method Update
        y_values[i] = y_values[i-1] + h * v_values[i-1]
        v_values[i] = v_values[i-1] + h * (-k * y_values[i-1])
    
    # Plotting Position (Solid)
    plt.plot(x_values, y_values, label=f'y(x) h={h}, k={k}')
    # Plotting Velocity (Dashed) - using the same color as the position line
    color = plt.gca().lines[-1].get_color() 
    plt.plot(x_values, v_values, color=color, linestyle='--', alpha=0.5, 
             label=f'v(x) h={h}, k={k}')

plt.xlabel('x')
plt.ylabel('Amplitude')
plt.title('Euler\'s Method: Position vs Velocity for varying h and k = 1.0')
# Move legend outside because there are now 10 lines
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize='small')
plt.grid(True)
plt.tight_layout()
plt.show()

# h values from the challenge with k=2.0
configs = [(.01, 2.0), (.02, 2.0), (.05, 2.0), (.1, 2.0), (.2, 2.0), (.3, 2.0), (.5, 2.0), (.01, 10.0)]
plt.figure(figsize=(14, 7))

for h, k in configs:
    x_values = np.arange(x_start, x_end + h, h)
    y_values = np.zeros(len(x_values))
    v_values = np.zeros(len(x_values))
    
    y_values[0] = y_0
    v_values[0] = v_0

    for i in range(1, len(x_values)):
        # Euler's Method Update
        y_values[i] = y_values[i-1] + h * v_values[i-1]
        v_values[i] = v_values[i-1] + h * (-k * y_values[i-1])
    
    # Plotting Position (Solid)
    plt.plot(x_values, y_values, label=f'y(x) h={h}, k={k}')
    # Plotting Velocity (Dashed) - using the same color as the position line
    color = plt.gca().lines[-1].get_color() 
    plt.plot(x_values, v_values, color=color, linestyle='--', alpha=0.5, 
             label=f'v(x) h={h}, k={k}')

plt.xlabel('x')
plt.ylabel('Amplitude')
plt.title('Euler\'s Method: Position vs Velocity for varying h and k = 2.0')
# Move legend outside because there are now 10 lines
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize='small')
plt.grid(True)
plt.tight_layout()
plt.show()
```

The graph becomes unstable starting around h=.3, and becomes noticably unstable at h=.5, yes it is becoming unstable.
As the value of K increases, the amplitude of the graphs increases.
Euler's method would not be able to solve this with a larger h value, as it would become more unstable with a larger h value. We would need an even smaller h value to solve this.
