```python
import numpy as np
import matplotlib.pyplot as plt

#setup
f = lambda x: np.sin(x)
a, b = 0, 2 * np.pi # Interval
N = 50 # Number of sample points
# Create a uniform grid
x = np.linspace(a, b, N)
y = f(x)
# step size
h = (b-a) / (N-1)

#true derivative
true_deriv = np.cos(x)

#Central Difference Approximation
df_central = (f(x + h) - f(x - h)) / (2 * h)

#Forward Difference Approximation
df_forward = (f(x+h) - f(x)) / h

#Backward Difference Approximation
df_backward = (f(x) - f(x - h)) / h

#create plot
plt.figure()
plt.plot(x, true_deriv, label='True Derivative')
plt.plot(x, df_central, 'r.', label='Central Difference')
plt.plot(x, df_forward, 'g.', label='Forward Difference')
plt.plot(x, df_backward, 'b.', label='Backward Difference')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.plot()
```
