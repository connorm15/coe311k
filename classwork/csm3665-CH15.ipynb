```python
import numpy as np
import matplotlib.pyplot as plt

# Define parameters
h_values = np.arange(0.01, 1.05, 0.05)
x_start, x_end = 0, 2

# Function representing dy/dx = -2y
def f(x, y):
    return -2 * y

for h in h_values:
    num_steps = int((x_end - x_start) / h)
    x_values = np.linspace(x_start, x_end, num_steps + 1)
    y_values = np.zeros(num_steps + 1)

    # Initial condition
    y_values[0] = 1  # y(0) = 1
    
    # Implement Euler's method

    for i in range(num_steps):

        y_values[i + 1] = y_values[i] + h * f(x_values[i], y_values[i])
    
    plt.plot(x_values, y_values, label=f"h = {h:.2f}")
    
    # Calculate Total Accumulated Error (comparing to exact solution y = e^(-2x))
    y_exact = np.exp(-2 * x_values)
    total_error = np.sum(np.abs(y_exact - y_values))
    print(f"Step size h={h:.2f} | Total Accumulated Error: {total_error:.4f}")


# Plot the results

plt.xlabel('x')

plt.ylabel('y')

plt.title("Approximation of dy/dx = -2y using Euler's Method")

plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

plt.show()
```
