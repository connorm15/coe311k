```python
import numpy as np
import matplotlib.pyplot as plt

# Define parameters
h_values = [0.1, 0.5, 1.0, 1.5]
x_start, x_end = 0, 10

# Function representing dy/dx = -2y
def f(x, y):
    return -2 * y
#funciton for exact solution
def exact_sol(x):
    return np.exp(-2 * x)

plt.figure()

for h in h_values:
    num_steps = int((x_end - x_start) / h)
    x_values = np.linspace(x_start, x_end, num_steps + 1)
    y_values = np.zeros(num_steps + 1)

    # Initial condition
    y_values[0] = 1  # y(0) = 1

    # Implement Euler's method
    for i in range(num_steps):
        y_values[i + 1] = y_values[i] + h * f(x_values[i], y_values[i])
    
    #calcualte errors
    y_true = exact_sol(x_values)
    
    #local error
    local_error = np.abs(y_true - y_values)
    
    #global error
    global_error = np.sum(local_error)
    
    # Plot the results
    plt.plot(x_values, y_values, label=f" Aprroximation with h = {h:.2f}")
    plt.plot(x_values, local_error, label=f"Local Erorr (Total: {global_error:.2f})")
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title("Approximation of dy/dx = -2y using Euler's Method")
    plt.legend()
    plt.show()
```
