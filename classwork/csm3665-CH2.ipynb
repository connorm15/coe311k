```python
def decimal_to_binary(n):
    if n == 0:
        return "0"
    remainders = []
    while n>0:
        remainder = n%2
        remainders.append(str(remainder))
        n = n//2
    return "".join(remainders[::-1])
number = 2025
binary_result = decimal_to_binary(number)
print(f"The binary representation of {number} is: {binary_result}")
```
